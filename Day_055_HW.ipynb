- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans

%matplotlib inline
```

```python
estimators = [("k_means_3", KMeans(n_clusters=3)),
              ("k_means_6", KMeans(n_clusters=6)),
              ("k_means_9", KMeans(n_clusters=9)),
              ("k_means_12", KMeans(n_clusters=12)),
              ("k_means_init_1", KMeans(n_clusters=3, n_init=1, init="random")),
              ("k_means_init_3", KMeans(n_clusters=3, n_init=3, init="random")),
              ("k_means_init_5", KMeans(n_clusters=3, n_init=5, init="random"))]
```

```python
fignum = 1
titles = ["3_clusters", "6_clusters", "9_clusters", "12_clusters",
          "init_1", "init_3", "init5"]

for name, est in estimators:
    fig = plt.figure(fignum, figsize = (4, 3))
#     ax = Axes3D(fig, rect=[0, 0, .95, 1], elev = 50, azim=150)
    
    
    est.fit(X)
    labels = est.labels_
    
    plt.scatter(X[:, 0], X[:, 1], c = labels.astype(np.float), edgecolor = "k")
    
#     ax.w_xaxis.set_ticklabels([])
#     ax.w_yaxis.set_ticklabels([])
#     ax.w_zaxis.set_ticklabels([])
    plt.title(titles[fignum - 1])
#     ax.dist = 12
    fignum += 1
```
